```python
# Downloading the files

import sunpy.net.attrs as a
import astropy.units as u
from sunpy.net import Fido

# Time range for data search
start_time = "2024-05-14T04:00:00"
end_time = "2024-05-14T06:00:00"

# Defining wavelength (171  is commonly used)
wavelength = a.Wavelength(171 * u.angstrom)

# Search for AIA data
result = Fido.search(a.Time(start_time, end_time), 
                     a.Instrument("AIA"), 
                     wavelength)

files = Fido.fetch(result)

print("AIA data downloaded:", files)
```
